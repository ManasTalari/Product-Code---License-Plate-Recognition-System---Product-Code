```python
import sqlite3
import random
import string

# Function to generate random Indian name
def generate_indian_name():
    first_names = ['Ramesh', 'Suresh', 'Rajesh', 'Deepak', 'Amit', 'Sandeep', 'Neha', 'Pooja', 'Ritu', 'Priya', 'Anita', 'Sunita']
    last_names = ['Kumar', 'Singh', 'Sharma', 'Verma', 'Gupta', 'Yadav', 'Mishra', 'Patel', 'Jha', 'Chauhan']
    return random.choice(first_names) + " " + random.choice(last_names)

# Function to generate random Indian phone number
def generate_indian_phone_number():
    area_code = random.choice(['98', '92', '97', '96', '93', '81', '83', '80', '82', '88'])
    exchange_code = ''.join(random.choices(string.digits, k=3))
    subscriber_number = ''.join(random.choices(string.digits, k=4))
    return f'({area_code}) {exchange_code}-{subscriber_number}'

# Function to generate random Indian address
def generate_indian_address():
    cities = ['Mumbai', 'Delhi', 'Bangalore', 'Kolkata', 'Chennai', 'Hyderabad', 'Pune', 'Ahmedabad', 'Surat', 'Jaipur', 'Lucknow']
    states = ['Maharashtra', 'Delhi', 'Karnataka', 'West Bengal', 'Tamil Nadu', 'Telangana', 'Uttar Pradesh', 'Gujarat', 'Rajasthan']
    return random.choice(string.ascii_uppercase) + ''.join(random.choices(string.ascii_lowercase, k=random.randint(5, 15))) + f", {random.choice(cities)}, {random.choice(states)}"

# Generate random data and insert into database
conn = sqlite3.connect('indian_random_data.db')
cursor = conn.cursor()

# Create table
cursor.execute('''
    CREATE TABLE IF NOT EXISTS people (
        id INTEGER PRIMARY KEY,
        name TEXT,
        age INTEGER,
        license_plate TEXT UNIQUE,
        phone_number TEXT UNIQUE,
        address TEXT,
        citizen TEXT
    )
''')

# Generate and insert random data
for _ in range(50):
    name = generate_indian_name()
    age = random.randint(18, 80)
    license_plate = ''.join(random.choices(string.ascii_uppercase + string.digits, k=10))
    phone_number = generate_indian_phone_number()
    address = generate_indian_address()
    citizen = "Indian"

    try:
        cursor.execute('''
            INSERT INTO people (name, age, license_plate, phone_number, address, citizen)
            VALUES (?, ?, ?, ?, ?, ?)
        ''', (name, age, license_plate, phone_number, address, citizen))
    except sqlite3.IntegrityError:
        # Handle case where generated license plate or phone number is not unique
        pass

conn.commit()
conn.close()

print("Random Indian data inserted successfully!")
```
